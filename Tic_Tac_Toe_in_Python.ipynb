# Tic Tac Toe Game in Python

Below is the code for the game :-

```python
from IPython.display import clear_output

def display_board(board):
    clear_output()
    print('\n\t       |7|             |8|             |9|\n')
    print('\t\t \t|\t \t|\t')
    print('\t\t'+board[7]+'\t|\t'+board[8]+'\t|\t'+board[9])
    print('\t----------------+---------------+----------------')
    print('\t\t \t|\t \t|\t')
    print('|4|             '+board[4]+'\t|\t'+board[5]+'\t|\t'+board[6]+'            |6|')
    print('\t\t \t|\t \t|\t')
    print('\t----------------+---------------+----------------')
    print('\t\t'+board[1]+'\t|\t'+board[2]+'\t|\t'+board[3])
    print('\t\t \t|\t \t|\t')
    print('\n\t       |1|             |2|             |3|\n')
    
def player_input():
    marker=''
    while marker!='X' and marker!='O':
        marker=input('Do you want to play with X or O ?').upper()
    if marker=='X':
        player1='X'
        player2='O'
    else:
        player1='O'
        player2='X'
    return (player1,player2)

def place_marker(board, marker, position):
    board[position]=marker
    
def win_check(board, mark):
    if board[1]==board[2]==board[3]==mark:
        return True
    elif board[4]==board[5]==board[6]==mark:
        return True
    elif board[7]==board[8]==board[9]==mark:
        return True
    elif board[1]==board[5]==board[9]==mark:
        return True
    elif board[3]==board[5]==board[7]==mark:
        return True
    elif board[1]==board[4]==board[7]==mark:
        return True
    elif board[2]==board[5]==board[8]==mark:
        return True
    elif board[3]==board[6]==board[9]==mark:
        return True
    return False

from random import randint

def choose_first():
    return randint(1,2)

def space_check(board, position):
    if board[position]==' ':
        return True
    return False

def full_board_check(board):
    for i in board:
        if i==' ':
            return False
    return True

def player_choice(board):
    position=0
    while (position<1 or position>9) or (not space_check(board,position)):
        position=int(input('Where do you want to place your marker (1-9)?'))
    return position

def replay():
    choice='x'
    while choice!='Y' and choice!='N':
        choice=input('Do you want to play again (Y/N)?').upper()
    if choice=='Y':
        return True
    return False

import time
print('Welcome to Tic Tac Toe!')

while True:
    # Set the game up here
    player_board=['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    markers=player_input()
    print('\nPlayer 1: '+markers[0]+'\nPlayer 2: '+markers[1])
    choice1=choose_first()
    print('\n<--Player '+str(choice1)+' will play first-->')
    if choice1==1:
        choice2=2
    else:
        choice2=1
    print('Loading...')
    time.sleep(2) 

    while True:
        if full_board_check(player_board):
            print('Oops!!! This match is a tie......')
            break
        print('\n<--Player '+str(choice1)+'-->')
        position=player_choice(player_board)
        place_marker(player_board,markers[choice1-1],position)
        display_board(player_board)
        if win_check(player_board, markers[choice1-1]):
            print('Hurray!!! Player '+str(choice1)+' WON******')
            break
            
        if full_board_check(player_board):
            print('Oops!!! This match is a tie......')
            break
        print('\n<--Player '+str(choice2)+'-->')
        position=player_choice(player_board)
        place_marker(player_board,markers[choice2-1],position)
        display_board(player_board)
        if win_check(player_board, markers[choice1-1]):
            print('Hurray!!! Player '+str(choice2)+' WON******')
            break

    if not replay():
        break
```
